## Data Cleaning
- whenever you have a lot of missing record values in the given dataset then you can isnull()
- isnull() will notify the any missing values in the dataframe

```python
# Entire data set
df.isnull().sum()
```

```python
# Only column missing values
df["Season"].isnull().sum()
```

- **dropna()**  This function allows you to drop all or some rows that have missing values
- **fillna()**  This function allows you to replace the rows that have missing values with the value that you want to pass

```python
df.shape
```

```python
# Drop rows with 1 null values
new_df = df.dropna(axis=0,how="any")
```

## np.zeros() np.ones()
- To intialize the weigths during the first iteration of TensorFlow

```python
x = np.zeros((2,2))  # dtype as float by default
print(x)
```

```python
x = np.zeros((2,2),dtype=np.int16)   # dtype has changed to int
print(x)
```

```python
x = np.ones((3,3,))   
print(x)
```

```python
x = np.ones((3,3),dtype=np.int16)
print(x)
```

```python
x2 = np.arange(16)
print(x2)
```

```python
# Convert the 1-D into 2-D
x3 = x2.reshape(4,4)
print(x3)
```

```python
x3 = x3/4
print(x3)
```

## Visualizing Data
- an intersting way to display dataframe
- with matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color="green")
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.show()
```

```python
marks = [79,82,85,63]
avg_marks = [56,63,45,72]
divisions = [1,2,3,4]
index = np.arange(len(divisions))
width = 0.70
plt.bar(divisions,marks,color="blue",width=0.3)
plt.show()
```

```python
plt.barh(divisions,marks,color="skyblue")
plt.show()
```

```python
lang = ["Python","Java","C","C++"]
rating = [75,70,55,65]
plt.pie(rating,labels = lang,shadow=True,startangle=90,autopct="%1.1f%%")
plt.legend(loc="best")
plt.show()
```
